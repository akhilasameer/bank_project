<h4 style='color: red'>IDENTIFY THE OUTLIERS WITH IQR METHOD</h4>

```python
# Initialize a DataFrame to store outliers
outlier_results = pd.DataFrame()

# to detect outliers using IQR method
for feature in numerical_cols:
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Display outliers for the current feature
    outliers = data[(data[feature] < lower_bound) | (data[feature] > upper_bound)]
    outlier_results = pd.concat([outlier_results, outliers], ignore_index=True)

# Display all detected outliers
print("All detected outliers:",outlier_results.shape)
```

<h4 style='color:red'>TREAT THE OUTLIERS WITH CAPPING METHOD</h4>

```python
for feature in numerical_cols:
    lower_cap = data[feature].quantile(0.01)
    upper_cap = data[feature].quantile(0.99)
    outliers_before_capping = data[(data[feature] < lower_cap) | (data[feature] > upper_cap)]
# Cap the outliers
    data[feature] = data[feature].clip(lower=lower_cap, upper=upper_cap)
    
# Display the first few rows of the modified dataset
print(data.head())
```

```python
# Identify and remove capped outliers
cleaned_data = data[(data[feature] >= lower_cap) & (data[feature] <= upper_cap)]

# Display the shape of the cleaned dataset
print("Shape of cleaned data after removing capped outliers:", cleaned_data.shape)
```

<h3 style='color:green'>11. MODEL EVALUATION</h3>

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score

# Evaluation function
def evaluate_model(model_name, y_test, y_predict):
    # Calculate metrics
    acc = accuracy_score(y_test, y_predict)
    prec = precision_score(y_test, y_predict, average='weighted')
    rec = recall_score(y_test, y_predict, average='weighted')
    f1 = f1_score(y_test, y_predict, average='weighted')

    # Print metrics
    print(f"\nModel: {model_name}")
    print("Accuracy:", acc)
    print("Precision Score:", prec)
    print("Recall Score:", rec)
    print("F1 Score:", f1)

    # Confusion Matrix and Classification Report
    conf_matrix = confusion_matrix(y_test, y_predict)
    print(f"Confusion Matrix for {model_name}:\n{conf_matrix}")
    print(f"Classification Report for {model_name}:\n{classification_report(y_test, y_predict)}")

# Evaluate models
evaluate_model("Logistic Regression", y_test, y_pred_log)
evaluate_model("Random Forest", y_test, y_pred_rf)
evaluate_model("Decision Tree", y_test, y_pred_dt)
evaluate_model("KNN", y_test, y_pred_knn)
evaluate_model("Gradient Boosting", y_test, y_pred_gb)
```
